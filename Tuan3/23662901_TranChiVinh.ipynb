```python
# MA TRAN TUONG QUAN TUYEN TINH(PERSON) CUA CAC CAC
# QUANTITYORDERED,PRICEEACH,SALES
df[['QUANTITYORDERED','PRICEEACH','SALES']].corr()

# Mặc định .corr() trong pandas sẽ tính hệ số tương quan Pearson.
# Hệ số tương quan Pearson
# Giá trị nằm trong khoảng [-1, 1]:
# +1 → tương quan tuyến tính dương hoàn hảo
# 0 → không có tương quan tuyến tính
# -1 → tương quan tuyến tính âm hoàn hảo
```

```python
#hoat 
df[['QUANTITYORDERED','PRICEEACH','SALES']].cov()
 
# tính ma trận hiệp phương sai (Covariance matrix) thay vì ma trận tương quan.
# .cov() trong pandas tính hiệp phương sai giữa các cột số.
# Hiệp phương sai (Covariance) đo mức độ hai biến thay đổi cùng nhau, công thức:
```



Nếu:

Cov > 0 → khi X tăng thì Y cũng có xu hướng tăng (quan hệ cùng chiều).

Cov < 0 → khi X tăng thì Y có xu hướng giảm (quan hệ ngược chiều).

Cov ≈ 0 → không có mối quan hệ tuyến tính rõ ràng.

```python
# Viết hàm foo(x) nếu x<10 trả về Bad, 
# nếu x>=10 và <50 trả về Good, 
# ngược lại trả về Excellent
def foo(x):
    if x < 10:
        return 'BAD'
    elif x >= 10 and x < 50:
        return 'GOOD'
    else:
        return 'EXCELLENT'

# Áp dụng cho cột MONTH
df['MONTH'] = df[['QUANTITYORDERED']].applymap(foo)

# Hiển thị kết quả
df[['QUANTITYORDERED','MONTH']]
```

```python
# Viết hàm so sánh giá trị: nếu x <= y trả về YES, ngược lại NO
def ftrust(x, y):
    if x <= y:
        return 'YES'
    else:
        return 'NO'

# Áp dụng hàm để gán giá trị trả về cho TRUST
df['TRUST'] = list(map(ftrust, df['PRICEEACH'], df['MSRP']))
df[['PRICEEACH','MSRP','TRUST']]
```

```python
#Thay đổi giá trị cho cột QTR_ID là Q1 nếu giá trị là 1, Q2 nếu giá trị là 2,...
dict_map= {1:'Q1', 2:'Q2', 3:'Q3', 4:'Q4'}
df['QTR_ID']= df['QTR_ID'].map(dict_map)
```

```python
# XÁC ĐỊNH CÁC HÀM TỔNG HỢP TRÊN CÁC BIẾN ĐỊNH LƯỢNG VÀ ĐỊNH TÍNH
#Hiển thị giá trị tổng , giá trị nhỏ nhất cho cột QUANTITYORDERED, giá trị nhỏ nhất
#và lớn nhất và trung bình cho cột PRICEEACH, giá trị nhỏ nhất và trung bình cho cột SALES
df.aggregate({'QUANTITYORDERED':['sum','min'],'PRICEEACH':['min','max','mean'],'SALES':['min','mean']})
```

```python
# GROUP BY: NHÓM

#Hiển thị giá trị tổng , giá trị nhỏ nhất cho cột QUANTITYORDERED, giá trị nhỏ nhất
#và lớn nhất và trung bình cho cột PRICEEACH, giá trị nhỏ nhất và trung bình cho cột SALES
df.aggregate({'QUANTITYORDERED':['sum','min'],'PRICEEACH':['min','max','mean'],'SALES':['min','mean']})
```

```python
#Thống kê tổng, trung vị, độ lệch chuẩn của SALES theo từng nhóm DEALSIZE
df.groupby(['DEALSIZE'])['SALES'].agg(['sum','mean','std'])
```

```python
#Thống kê tổng, trung vị, độ lệch chuẩn của SALES theo từng nhóm DEALSIZE, cùng DEALSIZE thì nhóm theo QTR_ID
df.groupby(['DEALSIZE','QTR_ID'])['SALES'].agg(['sum','mean','std'])
```
